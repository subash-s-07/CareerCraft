```python
import asyncio
import aiohttp
from bs4 import BeautifulSoup

# Base LinkedIn Jobs Search URL
list_url = "https://www.linkedin.com/jobs/search?keywords=Data%20Analysis&location=India"

async def fetch(session, url):
    """Fetch content from a URL asynchronously."""
    async with session.get(url) as response:
        return await response.text()

async def extract_job_ids():
    """Extract job IDs from the search page."""
    async with aiohttp.ClientSession() as session:
        html = await fetch(session, list_url)
        soup = BeautifulSoup(html, "html.parser")
        job_list = soup.find('ul', class_='jobs-search__results-list')
        
        if not job_list:
            return []

        job_ids = []
        for job in job_list.find_all('li'):
            job_div = job.find('div', attrs={"data-entity-urn": True})
            if job_div:
                job_id = job_div["data-entity-urn"].split(":")[-1]
                job_ids.append(job_id)
        
        return job_ids[:5]  # Limit to first 5 for speed

async def fetch_job_descriptions(job_ids):
    """Fetch job descriptions asynchronously."""
    job_hash = {'Job_ID': [], 'Role': [], 'Seniority': [], 'Emp_Type': [], 'Job function': [], 'Industries': [], 'Job_Des': [], 'Link': []}

    async with aiohttp.ClientSession() as session:
        tasks = [fetch(session, f'https://www.linkedin.com/jobs/view/{job_id}/') for job_id in job_ids]
        responses = await asyncio.gather(*tasks)

        for job_id, html in zip(job_ids, responses):
            soup = BeautifulSoup(html, "html.parser")

            try:
                job_des = soup.find('div', class_='show-more-less-html__markup')
                job_des_text = job_des.text.strip() if job_des else "N/A"

                role = soup.find('h1', class_='top-card-layout__title')
                role_text = role.text.strip() if role else "N/A"

                criteria = soup.find_all('span', class_='description__job-criteria-text')

                job_hash['Job_ID'].append(job_id)
                job_hash['Role'].append(role_text)
                job_hash['Seniority'].append(criteria[0].text.strip() if len(criteria) > 0 else "N/A")
                job_hash['Emp_Type'].append(criteria[1].text.strip() if len(criteria) > 1 else "N/A")
                job_hash['Job function'].append(criteria[2].text.strip() if len(criteria) > 2 else "N/A")
                job_hash['Industries'].append(criteria[3].text.strip() if len(criteria) > 3 else "N/A")
                job_hash['Job_Des'].append(job_des_text)
                job_hash['Link'].append(f'https://www.linkedin.com/jobs/view/{job_id}/')

            except Exception as e:
                print(f"Error parsing job {job_id}: {e}")

    return job_hash

async def main():
    job_ids = await extract_job_ids()
    job_data = await fetch_job_descriptions(job_ids)
    print(job_data)
    pd.DataFrame(job_data)

# Run the async functions
asyncio.run(main())
```
